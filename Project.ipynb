```python
def rep_xxk_w_0s(tokens):
    ntokens = []
    
    for token in tokens:
        matched_tok = re.match(r'^(\d+)k$', token,re.IGNORECASE)
        if matched_tok:
            val = int(matched_tok.group(1)) * 1000
            ntokens.append(str(val))
        else:
            ntokens.append(token)
    
    return ntokens
test = 'I need 50K wale  phones'
rep_xxk_w_0s(word_tokenize(test))
```

```python
def contains_dig(tokens):
    str_with_dig=[]
    for token in tokens:
        if isinstance(token, str) and re.fullmatch(r'\d+', token):
            str_with_dig.append(token)
    if len(str_with_dig) > 0:
        return str_with_dig
    else:
        return []
```
